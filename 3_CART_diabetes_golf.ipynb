```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./diabetes.csv')
#df = pd.read_csv('./golf_dataset_the_origin.csv')
```

```python
print(df.columns)


X = df.drop('Outcome', axis=1)           #use for 'diabetes' dataset
y = df['Outcome']                        #use for 'diabetes' dataset


#X = df.drop('Play', axis=1)             #use for 'golf' dataset and here 'play' is the class label given in question if not given no need to write these 2 lines
#y = df['Play']                          #use for 'golf' dataset
```

```python
model = DecisionTreeClassifier(criterion='gini')
```

```python
X_encoded = pd.get_dummies(X)

model.fit(X_encoded, y)

```

```python
plt.figure(figsize=(160, 120))
tree.plot_tree(
    model,
    feature_names=X.columns,
    class_names=['No', 'Yes'],
    filled=True
)
plt.show()
```

Short Note on CART

•CART stands for Classification and Regression Trees.
•It was introduced by Breiman, Friedman, Olshen, and Stone.
•CART is a decision tree algorithm used for both:
•Classification problems
•Regression problems
•It uses Gini Index for classification.
•It uses Mean Squared Error (MSE) for regression.
•CART produces a binary decision tree.
•It can handle both categorical and continuous attributes.
•CART follows a top-down, greedy approach.

Gini Index
Measures impurity of a dataset. Gini(D)=1−∑pi^2​
Ginia(D)=∑j=1 to v |Dj|/|D|*Gini(D)

•Algorithm for CART (Point Wise)

Input:

Training dataset with attributes and target variable

Output:

Binary decision tree

Steps:

1.Start with the complete training dataset at the root node.
2.For each attribute, consider all possible binary splits.
3.Compute the Gini Index for each split (for classification).
4.Select the split with the minimum Gini Index.
5.Create a decision node based on the selected split.
6.Divide the dataset into two subsets.
7.Repeat steps 2 to 6 recursively for each child node.
8.Stop splitting when:

   All records belong to the same class, or

   No further improvement is possible, or

  Minimum node size is reached.

9.Assign the class label (or mean value for regression) at leaf nodes.

•Advantages of CART

1.Works with both continuous and categorical data.

2.Produces binary trees, which are easy to evaluate.

3.Can handle missing values.


•Limitations of CART

Can overfit if tree grows too deep.

Binary splits may increase tree depth.

Sensitive to small variations in data.
