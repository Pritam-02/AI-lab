```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('./iris.csv')
#df = pd.read_csv('./golf_dataset_the_origin.csv')
```

```python
X = df.drop('species', axis=1)    #use for 'iris' dataset
y = df['species']                 #use for 'iris' dataset


#y = df['Play']                     #use for 'golf' dataset
#X = df.drop('Play', axis=1)        #use for 'golf' dataset

le=LabelEncoder()
y_encoded = le.fit_transform(y)

#X['Outlook'] = le.fit_transform(X['Outlook']) #use for 'golf' dataset
```

```python
model = DecisionTreeClassifier(criterion='entropy')

model.fit(X, y_encoded)
```

```python
plt.figure(figsize=(26, 19))
plot_tree(
    model, feature_names=X.columns, class_names=[
        'virginica', 'setosa', 'versicolor'],
    filled=True
)
plt.show()
```

```python
m2 = GaussianNB()
m2.fit(X, y_encoded)
```

```python
ns = pd.DataFrame([
    [5.3, 3.6, 1.6, 0.6]
],
columns=X.columns)

pred = m2.predict(ns)
res = le.inverse_transform(pred)
print(res[0])
```

Short Note on ID3

•ID3 stands for Iterative Dichotomiser 3.
•It is a decision tree learning algorithm proposed by J. R. Quinlan.
•ID3 is based on Information Theory.
•It uses Entropy and Information Gain to select the best attribute for splitting data.
•The attribute with the highest information gain becomes the decision node.
•It works only with categorical attributes.
•ID3 creates the tree in a top-down and greedy manner.
•It is mainly used for classification problems.

Algorithm for ID3 (Point Wise)
•Input:Training dataset with attributes and class labels

•Output: Decision tree

•Steps:
1.Calculate the entropy of the entire dataset.

2.or each attribute, calculate the information gain.

3.Select the attribute with the highest information gain.

4.Create a decision node for the selected attribute.

5.Split the dataset based on attribute values.

6.Repeat steps 1 to 5 recursively for each subset.

7.Stop recursion when:

   •All instances belong to the same class, or

   •No attributes are left for further splitting.

8.Assign the majority class as the leaf node if stopping condition is met.

•Advantages of ID3:
1.Simple and easy to understand.

2.Produces readable decision trees.

3.Uses a strong theoretical base from information theory.

•Limitations of ID3:
1.Cannot handle continuous attributes directly.

2.Sensitive to noisy data.

3.May lead to overfitting.
