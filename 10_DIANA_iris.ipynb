```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
# iris = load_iris()
# X = iris.data    # features only



df = pd.read_csv('./iris.csv')

X = df.drop('species', axis = 1)
y = df['species']
```

Single linkage

```python
Z_single = linkage(X, method='single', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_single)
plt.title("Single Linkage Clustering (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

Complete linkage

```python
Z_complete = linkage(X, method='complete', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_complete)
plt.title("Complete Linkage Clustering (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

Avg linkage

```python
Z_average = linkage(X, method='average', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_average)
plt.title("Average Linkage Clustering (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

Cut the dendogram into 3 clusters

```python
clusters = fcluster(Z_average, t=3, criterion='maxclust')
print(clusters)
```
