```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
iris = load_iris()
X = iris.data
```

Single linkage

```python
Z_single = linkage(X, method='single', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_single)
plt.title("AGNES - Single Linkage (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

Complete

```python
Z_complete = linkage(X, method='complete', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_complete)
plt.title("AGNES - Complete Linkage (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

Avg.

```python
Z_average = linkage(X, method='average', metric='euclidean')

plt.figure(figsize=(10, 5))
dendrogram(Z_average)
plt.title("AGNES - Average Linkage (Iris)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

From clusters

```python
clusters = fcluster(Z_average, t=3, criterion='maxclust')
print(clusters)
```

HAC is a bottom up clustering where each data point starts as its own cluster and clusters are merged step by step based on their similarity untill only one cluster remains as a desired no. of clusters are reached
